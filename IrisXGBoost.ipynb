```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,GradientBoostingRegressor,AdaBoostClassifier, StackingClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
from sklearn import svm
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
x=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
gbr = GradientBoostingRegressor(max_depth=3,
                                n_estimators=80,
                                random_state=42)

gbr.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score,accuracy_score
```

```python
y_pred = gbr.predict(x_test)
r2_score(y_test, y_pred)
```

```python
# adaboosting
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier(n_estimators=100, random_state=42)
ada.fit(x_train, y_train)
```

```python
accuracy_score(y_test,ada.predict(x_test))
```

```python
estimators = [('rf', RandomForestClassifier(n_estimators=20)),
              ('gbc', GradientBoostingClassifier()),('knn', KNeighborsClassifier(n_neighbors=10)),('svr',svm.SVC(kernel='linear',random_state=0))]
clf = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict=clf.predict(x_test)
accuracy_score(y_test,y_predict)
```
